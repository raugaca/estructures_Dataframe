## Columnes seleccionades:

* **Month**: 1-12
* **UniqueCarrier**: Codi del Carrier
* **FlightNum**: número del vol

* **AirTime**: durada del vol en minuts
* **ArrDelay**: retràs d'arribada, en minuts. Un vol es considera que ha arribat a temps si opera menys de 15 minuts més tard del temps programat que es mostra als carries' Computerized Reservations Systems (CRS).
    * Es pot calcular com aquesta suma: CarrierDelay + WeatherDelay + NASDelay + SecurityDelay + LateAircraftDelay
    * També es pot calcular com la resta: ArrTime - CRSArrTime quan les altres columnes estàn vuides però això és degut a vols cancel.lats o desviats.

* **DepDelay**: retràs de sortida, in minutes
* **Origin**: codi IATA de l'aeroport d'origen
* **Dest**: codi IATA de l'aeroport de destí
* **Distance**: en milles

* **Cancelled**: vol cancel.lat (0/1) --> informació reduntant que es pot extreure del CancellationCode
* **CancellationCode**: motiu de la cancel.lació (A = carrier, B = climatologia, C = National Airspace System (NAS), D = seguretat)

* **Diverted**: vol desviat --> 1 = yes, 0 = no
* **CarrierDelay**: en minuts. Enrederiment de companyia. Per exemple: neteja de l'avió, reparació de danys de l'avió, espera per passatjers o tripulants de l'avió, equipatje, inspeccions de l'avió,...
* **WeatherDelay**: en minuts. Enredariments deguts a condicions meteorològiques complicades, que poden afectar la sortida, la ruta del vol o l'arribada.
* **NASDelay** en minuts. Enredariments controlats per la National Airspace System (NAS), pot incloure: condicions meteorològiques no extremes, operacions de l'aeroport, volum elevat de tràfic aeri, control de tràfic aèri,... etc.
* **SecurityDelay** en minuts: Enrederiment causat per evacuació de la terminal, concurse Security delay is caused by evacuation of a terminal or concurrència, reembarcament de l'avió per motius de seguretat,...
* **LateAircraftDelay** en minuts. Enredariments deguts a enrederiments del mateix avió amb vols anteriors que es van arrossegant durant el dia.

## Columnes descartades:
* **Year**: 2008 -->Totes les dades són del 2008, per tant no és una dada rellevant
* **DayOfWeek**: 1 (Dilluns) - 7 (Diumenge) --> no fa cap aportació rellevant
* **DayofMonth**: 1-31 --> en principi no aporta cap informació destacable pel seu anàlisis
* **ArrTime**: hora real d'arribada del vol (hora local - hhmm) --> Informació innecesària o redundant
* **DepTime**: hora real de sortida del vol (hora local - hhmm) --> Informació innecesària o redundant
* **CRSDepTime**: hora planificada de sortida del vol (hora local - hhmm) --> Informació innecesària o redundant
* **CRSArrTime**: hora planificada d'arribada del vol (hora local - hhmm) --> Informació innecesària o redundant
* **TailNum**: matrícula del avió --> Informació innecesàrea per fer l'estudi
* **TaxiIn**: taxi in time, en minuts (deduixo que és el temps per omplir l'avió de passatgers) --> Informació poc interessant per fer l'estudi
* **TaxiOut**: taxi out time en minuts (deduixo que és el temps per buidar l'avió de passatgers)--> Informació poc interessant per fer l'estudi
* **ActualElapsedTime**: en minuts, és el temps real entre la sortida i l'arribada --> AirTime+TaxiIn+TaxiOut, però també és igual al ArrTime-DepTime.
* **CRSElapsedTime**: en minuts , és el temps teóric entre la sortida i l'arribada --> AirTime+TaxiIn+TaxiOut amb valors teórics que no surten a la taula, però també és igual al CRSArrTime-CRSDepTime que sí que surten a la taula.

* Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).

```python
df_retallat['VelocitatMitjana']=df_retallat['Distance']/(df_retallat['AirTime']/60) # Velocitat mitja calculada en milles/hora
```

```python
def funcioEnderrerit(x):
    if x == 0:
        return 0
    elif x<0:
        return 1
    else:
        return 2
```

```python
df_retallat['VolEnrederit']=df_retallat["ArrDelay"].apply(funcioEnderrerit) # Vol en hora (0), vol avançat (1) i vol endarrerit (2) 
```

```python
df_retallat['RutaUnidireccional']=df_retallat["Origin"]+'-'+df_retallat["Dest"]
```

```python
def funcioRuta(x,y):#comparamos el origen y el destino y el que sea mayor de los dos 
#lo ponemos en primer lugar y el menor en segundo lugar
    a=""
    if x <y:
        a= x+"-"+y
    else:
        a= y+"-"+x
    return a
```

```python
df_retallat['RutaBidereccional']= df_retallat.apply(lambda x: funcioRuta(x["Origin"], x["Dest"]), axis = 1)
```

```python
print(df_retallat[['VelocitatMitjana','VolEnrederit','RutaUnidireccional','RutaBidereccional']].head())
```

* Fes una taula de les aerolínies amb més endarreriments acumulats.

```python
df_endarrer=df_retallat[df_retallat['VolEnrederit'] == 2] #Només considerem els vols endarrerits per evitar que els vols avançats 
# compensin part del resultat.
```

```python
df_endarrer[['UniqueCarrier', 'ArrDelay']].groupby('UniqueCarrier').sum().sort_values(by='ArrDelay', ascending=False).head(10)
```

* Quins són els vols més llargs? I els més endarrerits? Busca les rutes més llargues i les que acumulen més retards.

```python
#Top 10 de vols més llargs en distància
df_retallat[['FlightNum', 'Distance']].sort_values(by='Distance', ascending=False).drop_duplicates().head(10) 
```

```python
#Top 10 de vols amb enrederiments més llargs en minuts
df_retallat[['FlightNum', 'ArrDelay']].sort_values(by='ArrDelay', ascending=False).head(10) 
```

```python
#Top 10 de rutes més llargues en distància 
#(eliminem els duplicats en distància per treure la duplicitat del vol d'anada amb el de tornada)
df_retallat[['RutaBidereccional', 'Distance']].sort_values(by='Distance', ascending=False).drop_duplicates().head(10) 
```

```python
#Top 10 de rutes amb més endarreriments 
#Només considerem els vols endarrerits per evitar que els vols avançats compensin part del resultat.
df_endarrer2=df_retallat[df_retallat['VolEnrederit'] == 2] 
df_endarrer2[['RutaBidereccional', 'ArrDelay']].groupby('RutaBidereccional').sum().sort_values(by='ArrDelay', ascending=False).head(10)
```

* Aporta allò que consideris rellevant.

```python
# Top 10 d'aeroports de destí amb més desviacions de vols
df_retallat[['Dest', 'Diverted']].groupby('Dest').sum().sort_values(by='Diverted', ascending=False).head(10)
```

```python
# Top 10 d'aeroports de destí amb més cancelació de vols
df_retallat[['Dest', 'Cancelled']].groupby('Dest').sum().sort_values(by='Cancelled', ascending=False).head(10)
```

```python
# Top 10 de vols cancelats en funció de l'aerolinea i del codi de cancel-lació
df_cancel=df_retallat[df_retallat['CancellationCode'] != 'N'] 
df_cancel=df_cancel[['UniqueCarrier','CancellationCode']]
pivotTable = df_cancel.pivot_table(index= "UniqueCarrier", columns= "CancellationCode", values="CancellationCode", 
aggfunc= 'size', fill_value=0)
```

```python
pivotTable['Total']=pivotTable['A']+pivotTable['B']+pivotTable['C']
print(pivotTable.sort_values(by='Total', ascending=False).head(10))
#(A = carrier, B = climatologia, C = National Airspace System (NAS), D = seguretat)
```

```python
# Cancel.lacions en funció del mes
df_cancellation_month=df_retallat[['Month', 'Cancelled']].groupby('Month').agg([np.sum, np.size]).sort_values(by=[('Cancelled', 'sum')], ascending=False)
df_cancellation_month.head(5)
```

```python
# Ratio de vols per cada vol desviat per mes
df_diverted_month=df_retallat[['Month', 'Diverted']].groupby('Month').agg([np.sum, np.size])
df_diverted_month[('Diverted', 'Ratio')]=df_diverted_month[('Diverted', 'size')]/df_diverted_month[('Diverted', 'sum')]

df_diverted_month.sort_values(by=[('Diverted', 'Ratio')]).round(2)
```

```python
# Top 10 d'enredariment per aerolinea desglossat per tipus d'endarreriment
df_delays=df_retallat[['UniqueCarrier', 'CarrierDelay','WeatherDelay','NASDelay', 'SecurityDelay','LateAircraftDelay', 'ArrDelay']].dropna()
df_delays.groupby('UniqueCarrier').sum().sort_values(by='ArrDelay', ascending=False).head(10)
```
